```python
import random
import string

def show_title():
    print("=" * 40)
    print("        PASSWORD GENERATOR")
    print("=" * 40)

def get_password_length():
    while True:
        length = int(input("Enter password length: "))
        if length <= 0:
            print("Password length must be greater than 0.")
        else:
            return length

def get_character_preferences():
    use_lower = input("Include lowercase letters? (y/n): ").lower()
    use_upper = input("Include uppercase letters? (y/n): ").lower()
    use_digits = input("Include numbers? (y/n): ").lower()
    use_symbols = input("Include special characters? (y/n): ").lower()

    characters = ""

    if use_lower == 'y':
        characters += string.ascii_lowercase
    if use_upper == 'y':
        characters += string.ascii_uppercase
    if use_digits == 'y':
        characters += string.digits
    if use_symbols == 'y':
        characters += string.punctuation

    return characters

def generate_password(length, characters):
    password = ""
    for i in range(length):
        password += random.choice(characters)
    return password

def main():
    show_title()

    length = get_password_length()
    characters = get_character_preferences()

    if characters == "":
        print("No character set selected. Cannot generate password.")
    else:
        password = generate_password(length, characters)
        print("\nPassword generated successfully!")
        print("Your password:", password)

main()
```
