```python
print("===================================")
print("        BMI CALCULATOR PROGRAM      ")
print("===================================")

name = input("Enter your name: ")

weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))

if weight <= 0 or height <= 0:
    print("Invalid input! Weight and height must be positive values.")
else:
    bmi = weight / (height * height)

    print("\nHello", name)
    print("Your weight:", weight, "kg")
    print("Your height:", height, "m")
    print("Your BMI is:", round(bmi, 2))

    if bmi < 18.5:
        print("BMI Category: Underweight")
        print("Advice: You may need to gain some weight.")
    elif bmi >= 18.5 and bmi < 25:
        print("BMI Category: Normal weight")
        print("Advice: Keep maintaining your healthy lifestyle.")
    elif bmi >= 25 and bmi < 30:
        print("BMI Category: Overweight")
        print("Advice: Regular exercise is recommended.")
    else:
        print("BMI Category: Obese")
        print("Advice: Please consult a healthcare professional.")

print("\nThank you for using the BMI Calculator!")
```
